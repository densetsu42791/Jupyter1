```python
import numpy as np
import matplotlib.pyplot as plt


def func(x):
    return x**3 - 6 * x**2 + 10

def func_derivative(x):
    return 3 * x**2 - 12 * x - 15


x = np.linspace(-5, 10, 150)


def descent_and_plot(X_start=10, learning_rate=0.01, thresholod=0.0001):
    X_start = X_start
    learning_rate = learning_rate
    thresholod = thresholod
    
    X_next = X_start - learning_rate * func_derivative(X_start)
    
    fig = plt.figure()
    fig.set_size_inches(10, 8)

    plt.plot(x, func(x), alpha=0.5, color='blue')
    plt.scatter([X_start, X_next], [func(X_start), func(X_next)],
               color='green', s=80)
    plt.plot([X_start, X_next], [func(X_start), func(X_next)],
               color='red', linewidth=3)

    plt.ylabel('f')
    plt.xlabel('x')

    
    while abs(X_start - X_next) > thresholod:
        X_start = X_next
        X_next = X_start - learning_rate * func_derivative(X_start)
        
        plt.scatter([X_start, X_next], [func(X_start), func(X_next)],
               color='green', s=20)
        plt.plot([X_start, X_next], [func(X_start), func(X_next)],
               color='red', linewidth=2)

    
    print(f'Найденная точка в конце градиентного спуска это X={X_next}')



descent_and_plot()
```
